```python
import os
from PIL import Image

input_dir = "//content/Delta Headlines"
output_dir = "/content/data/data"

os.makedirs(output_dir, exist_ok=True)

jpg_files = sorted([f for f in os.listdir(input_dir) if f.endswith('.jpg')])

for i, jpg in enumerate(jpg_files, start=1):
    img_path = os.path.join(input_dir, jpg)
    pdf_path = os.path.join(output_dir, f"{i}.pdf")

    image = Image.open(img_path)
    image.convert('RGB').save(pdf_path)

    print(f"Saved: {pdf_path}")

print("Conversioncompleted!")
```

```python
from llama_index import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
from llama_index.llms import HuggingFaceLLM
from llama_index.prompts.prompts import SimpleInputPrompt
```

```python
documents=SimpleDirectoryReader("/content/data/data").load_data()
documents
```

```python
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided.
"""

query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B")
model = AutoModelForCausalLM.from_pretrained("deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B")
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name='deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B',
    model_name="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={
        "torch_dtype": torch.float16,
        "load_in_8bit": True
    }
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index import ServiceContext
from llama_index.embeddings import LangchainEmbedding

embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
service_context
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
index
```
